```python
#Creates a two column matrix containing true values in column one and observed values in column two
import numpy as np
import random
k = 2
m = 1000
n = 5
e = 0.2
randmatrix = [[random.randrange(1,n) for i in range(k-1)] for j in range(m)]
randmatrix
```

```python
k-1 == k w/ prob e
otherwise random.randrange(1,n) == k w/ prob e/(n-1)
```

```python
choose a random number from 0 to 1 
if number is between 0 and 1-e, k == k-1
else k == random.uniform(0, e / n-1)
```

```python
matrix = []
for j in range(m):
    row = []
    for i in range(k):
        p = random.uniform(0,1)
        #print(p)
        if i == 0 : 
            row.append(random.randrange(0,n))
        elif 0 <= p <= 1- e: #0<= p <= 1-e
            row.append(row[i-1])
        else:
            row.append(random.randrange(1,n)) #want to set frequencies
    matrix.append(row)
matrix
```

```python
#datamatrix = []
datamatrix = [[0 for i in range(n)] for j in range(m)]
for j in range(m):
    for i in range(n):
        if i == matrix[j][1]:
            datamatrix[j][i] = 1-e
        else:
            datamatrix[j][i] = e/(n-1)
datamatrix
```
